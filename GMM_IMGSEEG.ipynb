```python
#Gausian Mixtrue Model(GMM)

import numpy as np
import cv2
```

```python
img = cv2.imread('NORMAL.png')
print(img.shape)
```

```python
img2 = img.reshape((-1, 3))

print(img2.shape)
```

```python
from sklearn.mixture import GaussianMixture as GMM
import matplotlib.pyplot as plt

gmm_model = GMM(n_components=3, covariance_type='full').fit(img2)
gmm_labels = gmm_model.predict(img2)

original_shape = img.shape
segmented = gmm_labels.reshape(original_shape[0], original_shape[1])


plt.imshow(segmented, cmap='gray')
plt.show()
```

```python
#how to choose the number of components

import numpy as np
import cv2
from sklearn.mixture import GaussianMixture as GMM
import matplotlib.pyplot as plt



img = cv2.imread('Covid (2).png')
img2 = img.reshape((-1, 3))


n_components = np.arange(1, 10)
gmm_models = [GMM(n, covariance_type='tied').fit(img2) for n in n_components]


plt.plot(n_components, [m.bic(img2) for m in gmm_models], label = 'BIC')
plt.xlabel('n_components')
```
